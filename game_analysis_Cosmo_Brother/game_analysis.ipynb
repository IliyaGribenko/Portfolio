## Описание и цели проекта

**Описание**

В проекте идет работа с данными о событиях, совершенных в мобильной `free-to-play` игре «Космические братья», находищийся на стадии `soft launch`. В ней пользователи строят свою космическую программу и пытаются преуспеть в нелёгком деле колонизации галактики. А также есть информация о стоимости и каналах привлечения пользователей.

В датасете представлены данные по игре пользователей на первом уровне. Завершение первого уровня требует от игрока выполнения одного из двух условий:
- победа над первым врагом (победа в сражении двух армий, базы игроков остаются целыми)
- реализация проекта - разработка орбитальной сборки спутников

В датасете содержатся данные первых пользователей приложения — когорты пользователей, которые начали пользоваться приложением в период с 4 по 10 мая включительно.

Основная монетизация игры — только планируется. Но предполагается, что в приложении будет происходить показ рекламы на экране с выбором типа объекта для постройки. Прибыль с показа одного рекламного банера равна `0.07`.

Требования к размещению рекламы:
- формат - баннерная реклама
- реклама должна быть привязана к событиям, которые на данный момент уже логируются

**Цели**

- найти `узкие места` в `retention` первого уровня
- предложить несколько моделей монетизации, которые будут окупать затраты на маркетинг в течении первого уровня, при этом не отпугивая игроков слишком большим количеством рекламы
- сделать предложения по оптимизации каналов привлечения трафика

**Описание датасетов**

Датасет `game_actions.csv`:
- `event_datetime` - время события;
- `event` - одно из трех событий:
    1. `building` -  обьект построен;
    2. `finished_stage_1` - первый уровень завершен;
    3. `project` - проект завершен;
- `building_type` - один из трех видов здания;
    1. `assembly_shop` - сборочный цех;
    2. `spaceport` - космопорт
    3. `research_center` - исследовательский центр;
- `user_id` - идентификатор пользователя;
- `project_type` тип реализованного проекта;

Датасет `ad_costs.csv`:
- `day` - день, в который был совершен клик по обьявлению;
- `source` - источник трафика;
- `cost` - стоимость клика;

Датасет `user_source.csv`:
- `user_id` - идентификатор пользователя
- `source` - источник, с которого пришел пользователь, установивший приложение;

**Примечание!**

Игрок совершает свое первое действие в игре на следующий день, после того как перешел по рекламному обьявлению!

## Подготовка данных

### Категоризация игроков (pvp, pve, churn)

Добавим в датасет `action` колонку `player_category` со следующими значениями:
- `pvp` - если игрок переходит на следующий уровень победив соперника
- `pve` - если игрок переходит на следующий уровень выполнив проект
- `churn` - если игрок не заканчивает уровень

```python
# выделим id игроков, которые прошли первый уровень
player_finished_stage_1 = actions.query('event == "finished_stage_1"')['user_id'].drop_duplicates()
```

```python
# выделим id игроков, которые прошли первый уровень через реализацию проекта
pve_user = actions.query('event == "project"')['user_id'].drop_duplicates()
```

```python
def pve_pvp_churn(user_id):
    '''' 
    Функция возвращает три значения `pve`, `pvp` или `churn`,
    в зависимости от того, к какой категории принадлежит пользователь
    
    На вход подается уникальный идентификатор пользователя.
    Функция использует 2 обьекта типа Series из глобального пространства имен:
    - player_finished_stage_1 содержит user_id игроков прошедших первый уровень любым способом
    - pve_user содержит user_id игроков прошедших первый уровень выполнив проект
    '''
    if user_id in pve_user.values:
        return 'pve'
    elif user_id in player_finished_stage_1.values:
        return 'pvp'
    else:
        return 'churn'
```

```python
actions['player_category'] = actions['user_id'].apply(pve_pvp_churn)
```

```python
# посмотрим на получившийся результат
actions['player_category'].value_counts(dropna=False, normalize=True)
```

**Примечание**
- примерно половину событий совершили игроки ушедшие в отток
- pvp игроки около 30%, pve 20%

### Профиля игроков

**Создадим датасет с пользовательскими профилями `user_profiles`, в него будет входить:**
- `user_id`
- `player_category`
- `source`
- `cac` - customer acquisition cost

```python
# создадим датасет с id пользователей, категорией игрока, датой первой  и последней сессий
profiles = (actions.sort_values(by=['user_id', 'event_date'])
                  .groupby(['user_id', 'player_category'])
                  .agg({'event_date' : ['first', 'last']})
                  .reset_index()
           )
profiles.columns = ['user_id', 'player_category', 'first_day', 'last_day']
```

```python
# добавим к профилям игроков источник их привлечения
profiles = profiles.merge(source, on='user_id')
```

```python
# добавим в таблицу `costs` количество привлеченных пользователей
costs = (costs.merge(profiles.groupby(['first_day', 'source'])['user_id'].count().reset_index(),
                      left_on=['source', 'day'], right_on=['source', 'first_day'])
               .rename(columns={'user_id':'count_user'})
        )
```

```python
# рассчитаем cac
costs['cac'] = costs['cost'] /  costs['count_user']
```

```python
# добавим в таблицу `profiles` данные о стоимости привлечения каждого пользователя
profiles = profiles.merge(costs[['source', 'first_day', 'cac']], on=['source', 'first_day'])
```

```python
# посмотрим на получившийся датасет
profiles.head()
```

### Подготовка данных к когортному анализу

```python
# добавим к таблице действий информацию о категории игрока, дату и источник привлечения, сас
actions = actions.merge(profiles, on=['user_id', 'player_category'])
```

```python
# рассчитаем life time игроков
actions['life_time'] = (actions['event_date'] - actions['first_day']).dt.days
# добавим к каждому значению life_time единицу, чтобы первый день соответствовал числу 1
actions['life_time'] = actions['life_time'] + 1
```

```python
# добавим столбец `all_cohort`, в будущем он понадобится нам для агрегации данных
actions['all_cohort'] = True
```

```python
# добавим к игрокам метки, в какой день они ушли в отток
actions['churn'] = (actions['event_datetime'] == actions['last_day'])
actions.loc[actions['player_category']!="churn", 'churn'] = False
```

Классический retention в данной работе построить не получится, так как у нас нет сведений об удержании пользователей на втором и последующих уровнях. 

Но у нас есть данные о пользователях, которые не завершили первый уровень, и так как к концу исследуемого периода DAU равен 3, можно считать, что эти пользователи ушили в отток.

Таким образом мы можем рассчитать отток пользователей в течении первого уровня и обратную ему величину Rolling retention. **Обратим внимание, что данные параметры относятся только к первому уровню**, например rolling retention на 15 день говорит только об удержании на первом уровне, если игрок ушел в отток на втором уровне, он все равно будет считаться не ушедшим.

```python
def churn_func(actions, dimension):
    """ Функция принимает на вход датасет с сессиями `actions`, 
    
    Функция возвращает 2 датафрейма, со значениями churn и rolling_retention
    """
    # создадим таблицу, отображающую количество игроков ушедших в отток
    churn = actions.query('churn == True').pivot_table(index=dimension, columns='life_time',
                                                   aggfunc='nunique', values='user_id')
    # вычислим кумулятивную сумму игроков ушедших в отток по когортам
    churn = churn.cumsum(axis=1)

    # вычислим размеры когорт
    cohort_size = (actions.groupby(dimension)['user_id'].nunique().reset_index()
                      .rename(columns={'user_id':'cohort_size'}).set_index(dimension)
                  )
    
    # добавим в датасет размеры когорт
    churn = churn.merge(cohort_size, left_index=True, right_index=True)
    
    # посчитаем процент игроков ушедших в отток по когортам
    churn = churn.div(churn['cohort_size'], axis=0).drop(columns='cohort_size')
    
    # вычисляем rolling_retention
    rolling_retention = (1 - churn)
    
    return churn, rolling_retention
```

```python
churn, rolling_retention = churn_func(actions, 'source')
```

```python
fig, axes = plt.subplots(figsize=(15,7), nrows=1, ncols=2)

churn.T.plot(ax=axes[0])
axes[0].set_title('Churn (first_level)')
axes[0].set_xlabel('life_time')
axes[0].set_ylabel('% churn')

rolling_retention.T.plot(ax=axes[1])
axes[1].set_title('Rolling retention (first_level)')
axes[1].set_xlabel('life_time')
axes[1].set_ylabel('% rolling retention')

plt.show()
```

**Примечание**
- отток между разными каналами привлечения примерно равен и к концу первого уровня составляет около 58%

## Модели монетизации для 1 уровня

### Функции для оценки модели монетизации

```python
def get_ltv_roi_cac(actions, model, dimension=[]):
    """"Функция принимает на вход датасет с сессиями `actions` и рассчитанными на каждого 
    пользователя  life_time, cac. А также нужно передать список `dimension`, в котором 
    указано по каким столбцам агригировать данные. И столбец по которому считать прибыль
    от показа рекламы `model`.

    Функция возвращает значения ltv, roi, cac. 
    """
    
    # вычисляем размеры когорт
    cohort_size = actions.groupby(dimension)['user_id'].nunique()
    
    # находим сумму выручки с накоплением
    cum_arpu = actions.pivot_table(index=dimension, columns='life_time', 
                                   values=model, aggfunc='sum').fillna(0).cumsum(axis=1)
    
    # считаем LTV
    ltv = cum_arpu.div(cohort_size, axis=0)
    
    # считаем CAC
    cac = actions[['user_id', 'cac'] + dimension].drop_duplicates().groupby(dimension).agg({'cac' : 'mean'})
    
    # считаем ROI
    roi = ltv.div(cac['cac'], axis=0)
    
    return ltv, roi, cac
```

```python
def plot_ltv_roi_cac(ltv_cohort, cac_cohort, roi_cohort, roi, dimension, n_day):
    """"Функция принимает на вход значения динамики ltv/cac/roi по когортам,
    `n_day` - за какой день строить график динамики ltv/cac/roi по когортам
    и параметр `dimension`, по которому агрегировались данные
    
    Функция строит 4 графика: график динамики ltv/cac/roi по когортам за N-ный день,
    график динамики roi по life_time
    """
    
    plt.figure(figsize=(15,15))
    
    # Строим график динамики LTV по когортам за N-ный день
    ax = plt.subplot(2, 2, 1)
    ltv_cohort.pivot_table(index='first_day', columns=dimension, values=n_day, aggfunc='mean').plot(ax=ax)
    plt.title(f'Динамика LTV на {n_day} день')
    plt.xlabel('Дата привлечения')
    plt.ylabel('$', rotation=0, fontsize=15)
    plt.ylim(0, 1)
    plt.xticks(rotation=45)
    
    # Строим график динамики CAC по когортам за N-ный день
    ax1 = plt.subplot(2, 2, 2, sharey=ax)
    cac_cohort.reset_index().pivot_table(index='first_day', columns=dimension, values='cac', aggfunc='mean').plot(ax=ax1)
    plt.title(f'Динамика CAC на {n_day} день')
    plt.xlabel('Дата привлечения')
    plt.xticks(rotation=45)
    
    # Строим график динамики ROI по когортам за N-най день
    ax3 = plt.subplot(2, 2, 3)
    roi_cohort.pivot_table(index='first_day', columns=dimension, values=n_day, aggfunc='mean').plot(ax=ax3)
    plt.title(f'Динамика ROI на {n_day} день')
    plt.xlabel('Дата привлечения')
    plt.ylabel('%', rotation=0, fontsize=15)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.ylim(0)
    
    # Строим график ROI
    ax4 = plt.subplot(2, 2, 4, sharey=ax3)
    roi.T.plot(ax=ax4)
    plt.title('ROI')
    plt.xlabel('Lifetime')
    plt.xticks(rotation=45)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупоемости')
    plt.ylim(0)
    
    plt.tight_layout()
    plt.show()
```

```python
def unit_economic(actions, model, dimension, n_day):
    """"Функция последовательно вызывает 2 другие функции, а именно
    `get_ltv_roi_cac` и `plot_ltv_roi_caс`
    """
    
    ltv, roi, cac = get_ltv_roi_cac(actions, model, [dimension])
    ltv_cohort, roi_cohort, cac_cohort =  get_ltv_roi_cac(actions, model, ['first_day'] + [dimension])
    
    plot_ltv_roi_cac(ltv_cohort, cac_cohort, roi_cohort, roi, dimension, n_day)
```

- даты начала и конца привлечения игроков с разных источников совпадают: с `2020-05-04` по `2020-05-10`
- в датасете данные о 13576 пользователях

**Маркетинг**

- траты на маркетинг неравномерны, на привлечение первой когорты было потрачено половина бюджета
    - это одна из причин недостаточного количества игроков в некоторых когортах
    
**Помимо уменьшения затрат на маркетинг от когорты к когорте, также уменьшается и количество привлеченных пользователей. Так как игра многопользовательская, небольшое количество онлайн игроков может сделать недоступным часть функционала игры. Что в свою очередь может привести к оттоку пользователей.** Факты указывающие на правдоподобность данной гипотезы:
- `ltv` `pve игроков` менее чувствительно к уменьшению размеров когорт, чем `ltv` `pvp игроков`
- чем меньше в когорте игроков, тем больше в ней отток

**Анализ каналов привлечения**

- `ltv` на всех каналах привлечения примерно равен
- при этом `cac` `facebook` и `instagram` выше примерно в полтора раза чем у `yandex_direct` и `youtube`
    - это делает вторые два источника гораздо выгоднее в разрезе окупаемости на первом уровне
    - но стоит отметить, что отток у всех источников рекламы одинаковый, и нам не хватает данных о последующих уровнях, чтобы оценить `ltv` на всем периоде жизни клиента в продукте
        - то есть возможно, что игроки пришедшие с `facebook` и `instagramm` на следующих уровнях начнут приносить больший доход, чем игроки пришедшие с `yandex_direct` и `youtube`

**Рекомендации**

Лучше всего себя показала **когорта размером примерно 6_000 игроков**, я советую сделать такое количество игроков минимальным при наборе новых когорт. Можно экспериментировать с размером когорт, увеличивая их.

Когорты стоит формировать примерно равных размеров.

Несмотря на то, что игроки привлеченные с `facebook` и `instagram` не окупаются на первом уровне, я советую не исключать их полностью из трафика, чтобы понять, как данный сеймент игроков монетизируется на последующих уровнях. Отметим, что удержание по всем каналам привлечения одинаково. Но возможно частичное уменьшение бюджета. При этом нужно соблюдать баланс между потребностью получить окупаемость на данный момент и достаточным размером выборки для сревнения.

На `youtube_channel_reklama` потрачено примерно в 2 раза меньше чем на другие каналы, хотя он показывает наибольший **ROI**. Я бы советовал увеличить закупку трафика через этот канал привлечения.

**Особенности поведения игроков ушедших в отток**

Пик игроков уходящих в отток приходится на постройку четвертого - шестого зданий. Эту информацию стоит передать геймдезайнерам для поиска причины.

**Возможные причины**

- возможно при переходе на следующий уровень `pvp` игрок одерживает победу, а проигравший теряет часть мотивации продолжать, так как его армия разбита и уходит в отток
- возможно есть какая-то особенность механики игры, которая мешает игрокам преодолеть данный этап
- или можно попробовать упростить первый уровень, чтобы игрок одерживал победу над не сильным ботом вместо игрока, так у всех будет возможность без труда пройти цикл игры

**Сразу отиметим, что для правильного внедрения рекламы без негативного влияния на пользователей требуется A/B-тестирование!** Так что окончательный выбор модели монетизации должен приниматься на основе тестов.

Также при моделировании не был учтен негативный эффект от показа рекламы.

**Разработанная модель монетизации.**
- показываем **баннеры** на экране с выбором типа обьекта для постройки начиная с третьего события, которое совершил пользователь
- после окончания игроками уровня и изучения проекта тоже показываем баннер с рекламой
- модель окупается, `ROI` примерно равен 0
